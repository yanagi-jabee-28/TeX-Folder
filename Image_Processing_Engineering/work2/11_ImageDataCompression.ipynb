必要なモジュールのインポート・画像ファイルパスの指定

```python
#モジュールのインポート
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 共通のディレクトリパス
common_path = '/content/drive/MyDrive/img2025/image/'
```

# **JPEGによる非可逆圧縮**

```python
# 画像を読み込む
img = cv2.imread(common_path + 'color_image.png')

# 非可逆圧縮の品質を指定 値の範囲 : 0～100
quality = 100

# 圧縮後の画像の保存先パス
output_path = common_path + f'compressed_image_{quality}.jpg'

# JPEG圧縮を行う


# 圧縮後の画像を読み込む
compressed_img = cv2.imread(output_path)

#pltで表示するためにBGRをRGBに反転する
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
compressed_img = cv2.cvtColor(compressed_img, cv2.COLOR_BGR2RGB)

# 実行結果の表示
plt.figure(figsize=(6, 3))

# 元画像
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray', vmin=np.min(img),vmax=np.max(img))
plt.axis('off')

# 圧縮画像
plt.subplot(1, 2, 2)
plt.title('Compressed Image')
plt.imshow(compressed_img, cmap='gray',vmin=np.min(compressed_img),vmax=np.max(compressed_img))
plt.axis('off')

plt.tight_layout()
plt.show()
```
